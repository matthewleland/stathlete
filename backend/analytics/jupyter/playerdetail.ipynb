```python


# Get career points for Stephen Curry
curry_career = playercareerstats.PlayerCareerStats(player_id='201939')
curry_points = pd.DataFrame(curry_career.get_data_frames()[0][['PLAYER_AGE', 'PTS']])

# Get career points for LeBron James
james_career = playercareerstats.PlayerCareerStats(player_id='2544')
james_points = pd.DataFrame(james_career.get_data_frames()[0][['PLAYER_AGE', 'PTS']])

# Group points by age and sum over all seasons
curry_points = curry_points.groupby('PLAYER_AGE').sum()
james_points = james_points.groupby('PLAYER_AGE').sum()

# Merge data for both players
merged_points = pd.merge(curry_points, james_points, on='PLAYER_AGE', suffixes=('_Curry', '_James'))

# Create line plot
fig, ax = plt.subplots()
merged_points.plot(kind='line', ax=ax)
ax.set_xlabel('Age')
ax.set_ylabel('Points')
ax.set_title('Career Points Comparison: Stephen Curry vs. LeBron James')
plt.show()
```

```python
fig, ax = plt.subplots()
ax.scatter(merged_points.index, merged_points['PTS_Curry'], label='Stephen Curry')
ax.scatter(merged_points.index, merged_points['PTS_James'], label='LeBron James')
ax.set_xlabel('Age')
ax.set_ylabel('Points')
ax.set_title('Career Points Comparison: Stephen Curry vs. LeBron James')
ax.legend()
plt.show()
```
